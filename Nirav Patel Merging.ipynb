```python
customer_actions = customer_actions.drop(['Phone_Number', 'Log_Time', 'ImportedDate'], axis = 1)
```

```python
customer_actions.head()
```

```python
customer_actions['Call_Time_seconds'] = (customer_actions['Call_End'] - customer_actions['Call_Start']).dt.total_seconds()
```

```python
customer_actions.to_csv('Clean_ExportCustomerActions.csv', index = False)
```
